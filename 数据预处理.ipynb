```python
import pandas as pd
import numpy as np
from datetime import datetime
import time
```

```python
def convert2_unixtime(s):
    year_s, mon_s, day_s = s.split('-')
    dtime = datetime(int(year_s), int(mon_s), int(day_s))
    unix_time = time.mktime(dtime.timetuple())
    # print(unix_time)
    return int(unix_time)
```

```python
data = pd.read_csv("final_geo_df.csv")
data
```

```python
d = data['date']
# display_html
data['date'] = d.apply(convert2_unixtime)
data
# data = data.sort_values(by=['date'])
# data
# data['date'] = time.mktime(d.timetuple())
# data
```

```python
# 简单用pandas.get_dummis
def one_hot(df,colName):
    dummies = pd.get_dummies(df[colName],prefix=colName)
    res = pd.concat([df.drop(columns=[colName]),dummies],axis=1)
    return res
```

```python
# 始终按时间顺序排列
data = data.sort_values(by=['date'])
```

```python
data = one_hot(data,'neutral')
data 
```

```python
data = one_hot(data,'rest_days')
data
```

```python
# 标准化 (x - x.mean())/x.std()
# output.mean() =  0
# output.std() = 1
def normalization(df,colName):
    d = df[colName]
    d = (d - d.mean())/d.std()
    # print(d)
    return d
```

```python
# 值类型的特征标准化一下
data['rank_diff'] = normalization(data,'rank_diff')
data['rank_diff'] = normalization(data,'rank_diff')
data['point_diff'] = normalization(data,'point_diff')
data['point_increase_home'] = normalization(data,'point_increase_home')
data['point_increase_away'] = normalization(data,'point_increase_away')
data
```

```python
data.to_csv('2.csv',index=False)
```
